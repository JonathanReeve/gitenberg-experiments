```python
import pandas as pd
import re
import requests
import sqlite3
```

```python
df = pd.read_hdf('pgaus.hdf', 'pgaus')
```

```python
for i, row in df.iterrows(): 
    if type(row.Text) != str and type(row.HTML) != str: 
        if type(row.htmlURL) == str: 
            url = row.htmlURL.strip()
            r = requests.get(url)
            if r.status_code == 200: 
                text = r.text
                print('Getting HTML for row: ', i, end='\r')
                df.set_value(i, 'HTML', text)
            else: 
                print('Something went wrong. Code not 200.')
        else: 
            print('No HTML available for this text.')
```

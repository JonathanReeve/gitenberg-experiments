```python
grouped = list(df.groupby(by='author'))
```

```python
authorSizes = {author: len(data) for (author, data) in authorDict.items()}
```

```python
for author in sorted(authorSizes, key=authorSizes.get, reverse=True): 
    print(author, authorSizes[author])
```

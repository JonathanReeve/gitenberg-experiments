```python
def getPlaintextURL(formats): 
    # Prefer utf-8 to everything. 
    if 'text/plain; charset=utf-8' in formats.keys(): 
        return formats['text/plain; charset=utf-8']
    # Otherwise try to get 'text/plain', which is often utf-8 by default
    if 'text/plain' in formats.keys(): 
        return formats['text/plain']
    # Otherwise get any plaintext file. 
    for key in formats.keys(): 
        if key.startswith('text/plain'): 
            return formats[key]
    # I GUESS we'll accept HTML. Prefer utf-8. 
    if 'text/html; charset=utf-8' in formats.keys(): 
        return formats['text/html; charset=utf-8']
    # How about straight up 'text/html' (often utf-8)?. 
    if 'text/html' in formats.keys(): 
        return formats['text/html'] 
    # I guess we'll take any HTML at this point. 
    for key in formats.keys(): 
        if key.startswith('text/html'):
            return formats[key]
    # Couldn't find anything
    return None
```

```python
# Construct a dictionary with the form bookID: filename
pgFilenameDict = {}
for fn in pgFilenames: 
    bookID = fn.split('/')[-1].split('.')[0]
    if bookID.endswith('-8'): 
        bookID = bookID[:-2]
        pgFilenameDict[bookID] = fn
    elif 'readme' in bookID: 
        continue
    elif bookID.isnumeric() == False:
        continue
    else: 
        if bookID in pgFilenameDict: 
            # If it's already there, it's utf-8. Don't overwrite it. 
            continue
        else: 
            pgFilenameDict[bookID] = fn
```

```python
len(pgFilenameDict)
```

```python
# Get PG filenames for those that don't have GB filenames
for i, row in mergedWithFilenames.iterrows(): 
    if type(row['filename']) == float: # If NaN
        bookID = str(int(row['id']))
        if bookID in pgFilenameDict: 
#             print('using filename: ', pgFilenameDict[bookID])
            mergedWithFilenames.set_value(i, 'filename', pgFilenameDict[bookID])
```

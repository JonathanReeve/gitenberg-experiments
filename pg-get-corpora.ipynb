```python
import pandas as pd
import textacy
import gutenberg
from gutenberg.cleanup import strip_headers
```

```python
df = pd.read_hdf('pg-text-3-filenames.hdf')
```

```python
dickens = df[df['author'] == 'Dickens, Charles']
```

```python
dickens = singleAuthorDF(df, 'Dickens, Charles')
dickens
```

```python
def singleAuthorDF(df, author, lang="en"): 
    authorDF = df[df['author'] == author] 
    # Querying for lists is tricky. 
    authorLangDF = authorDF[authorDF.languages.apply(lambda x: x == ['en'])]
    withFilenameDF = authorLangDF[authorLangDF.filename.notnull()]
    return withFilenameDF

def getText(fn): 
    """ Used with df.apply() """
    with open(fn, errors='ignore') as f: 
        text = f.read() 
    return text
```

```python
dickens['text'] = dickens.filename.apply(getText)
```

```python
dickens['text'] = dickens['text'].apply(strip_headers)
```

```python
dickens.to_json('dickens.json', force_ascii=False, orient='records', lines=True)
```

```python
records = textacy.fileio.read_json_lines('dickens.json')
```

```python
text_stream, metadata_stream = textacy.fileio.split_record_fields(records, 'text')
```

```python
corpus = textacy.Corpus('en', texts=text_stream, metadatas=metadata_stream)
```

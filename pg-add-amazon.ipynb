```python
from amazon.api import AmazonAPI
# import bottlenose
from secrets import *
import pandas as pd
import beeprint
from lxml import etree
import time
```

```python
amazon = AmazonAPI(AMAZON_ACCESS_KEY, AMAZON_SECRET_KEY, AMAZON_ASSOC_TAG)
```

```python
df = pd.read_hdf('pg-text-6-amazon.hdf')
```

```python
df['am_cat_ancestors'] = df['am_cat_ancestors'].astype(list)
```

```python
df['am_editorial_reviews'] = df['am_editorial_reviews'].astype(list)
```

```python
start, end = 46847, 100000
for i, row in df.iterrows(): 
    if i >= start and i <= end:
        try: 
            author = row.author
            title = row.title
            print('---') 
            print('Looking up #%s: %s' % (i, row.title))
            print('By: ', row.author)
            try: 
                result = amazon.search_n(1, Author=author, Title=title, SearchIndex='Books')[0]
            except: 
                time.sleep(1)
                continue
            nodes = result.browse_nodes[0] # Get category
            ancestors = [a.name for a in nodes.ancestors]
            df.set_value(i, 'am_title', str(result.title))
            df.set_value(i, 'am_author', str(result.author))
            df.set_value(i, 'am_cat', str(nodes.name))
            df.set_value(i, 'am_cat_ancestors', ancestors)
            df.set_value(i, 'am_genre', result.genre)
            df.set_value(i, 'am_editorial_reviews', result.editorial_reviews)
            df.set_value(i, 'am_sales_rank', str(result.sales_rank))
            df.set_value(i, 'am_isbn', result.isbn)
            print('  Title: ', result.title)
            print('  Category: ', nodes.name)
            print('  Cat ancestors: ', ancestors)
            time.sleep(1)
        except Exception: 
            time.sleep(1)
            continue
            
```

```python
df.to_hdf('pg-text-6-amazon.hdf', 'pg')
```

```python
def getPath(bookID):
    filename = "%s.txt" % bookID
    print(bookID, end='\r')
    for path in allFiles: 
        pathFile = path.split('/')[-1]
#         print(pathFile)
        if pathFile == filename: 
            return path
        else: 
            continue
    return None
```

```python
import pandas as pd
import glob
import os
import gutenberg
from ast import literal_eval
```

```python
df = pd.read_hdf('pg-metadata-rdf.hdf', 'pg-metadata')
```

```python
df = df.T
```

```python
len(df)
```

```python
dfText = df.loc[df.type == 'Text']
```

```python
dfText.to_hdf('pg-text-0.hdf', 'pg')
```

```python
def hasPR(subjects): 
    return True if 'PR' in subjects else False

def isEnglish(langs): 
    if langs == None: 
        return False
    return True if 'en' in langs else False
```

```python
df['hasPR'] = df['LCC'].apply(hasPR)
df['isEng'] = df['language'].apply(isEnglish)
```

```python
prEng = df.loc[(df['hasPR'] == True) & (df['isEng'] == True)]
```

```python
len(prEng)
```

```python
repodir = '/run/media/jon/SAMSUNG/gitenberg'
```

```python
ids = list(prEng['id'])
```

```python
repos = os.listdir(repodir)
```

```python
len(repos)
```

```python
prEngRepos = [repo for repo in repos if int(repo.split('_')[-1]) in ids]
```

```python
len(prEngRepos)
```

```python
txtFiles = [glob.glob(repodir+'/'+repo+'/*.txt') for repo in prEngRepos]
```

```python
utfTextFiles = []
for pair in txtFiles: 
    for filename in pair: 
        name = filename.split('/')[-1]
        if '-' not in name: 
            utfTextFiles.append(filename)
```

```python
prEng
```

```python
gitenbergMeta = pd.read_csv('data/gitenberg-metadata.csv')
```

```python
gitenbergMeta = gitenbergMeta.rename(columns={'Unnamed: 0': 'id'})
```

```python
def getLCC(subjects):  if type(subjects) == str: 
        subjects = literal_eval(subjects)
    if type(subjects) == list: 
        for subject in subjects: 
            if subject.startswith('!lcc'): 
                return subject[5:]
```

```python
gitenbergMeta['lcc'] = gitenbergMeta['subjects'].apply(getLCC)
```

```python
prEnGitenberg = gitenbergMeta.loc[(gitenbergMeta['language'] == 'en') & (gitenbergMeta['lcc'] == 'PR')]
```

```python
merged = pd.merge(prEnGitenberg, prEng, on='id', how='outer', suffixes=('_prEnGit', '_prEng') )
```

```python
merged = pd.merge(dfText, gitenbergMeta, on='id', how='outer', suffixes=('_GB', '_PG') )
```

```python
merged.columns
```

```python
merged.to_hdf('pr-text-1-gb-merged.hdf', 'pg')
```

```python
textFileDict = {int(fn.split('/')[-1].split('.')[0]): fn for fn in utfTextFiles if fn.split('/')[-1].split('.')[0].isnumeric()}
```

```python
filenamesDF = pd.DataFrame(list(textFileDict.items()), columns=['id', 'filename'])
```

```python
filenamesDF
```

```python
mergedWithFilenames = pd.merge(merged, filenamesDF, on='id', how='outer')
```

```python
mergedWithFilenames.to_hdf('preng-with-filenames.hdf', 'preng')
```

```python
def getText(filename): 
#     print(filename)
    if type(filename) == str: 
        with open(filename, errors='ignore') as f: 
            out = f.read()
        return out
    else: 
        return None
```

```python
mergedWithFilenames['text'] = mergedWithFilenames['filename'].apply(getText)
```

```python
mergedWithFilenames.to_json('preng-with-files.json')
```

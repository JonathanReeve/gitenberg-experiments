```python
import pandas as pd
import re
import requests
```

```python
df = pd.read_csv('gutindex_aus2.txt', sep='\t')
```

```python
for i, row in df.iterrows(): 
    string = row.URLOther
    bracketed = re.findall('\[.*?\]', string)
#     print(i, bracketed)
    for bracket in bracketed: 
        if bracket.startswith('[Author') or bracket.lower().startswith('[full name'):
            author = bracket[bracket.find(':')+2:][:-1]
#             print('Author: ', author)
            df.set_value(i, 'Author', author)
        elif bracket.startswith('[Title') or bracket.startswith('[Full Title') or bracket.startswith('[Full title'): 
            title = bracket[bracket.find(':')+2:][:-1]
#             print('Title: ', title)
            df.set_value(i, 'Title', title)
        elif bracket.startswith('[Subtitle'): 
            subtitle = bracket[bracket.find(':')+2:][:-1]
#             print('Subtitle: ', subtitle)
            df.set_value(i, 'Subtitle', subtitle)
        else: 
#             print('Notes: ', bracket)
            df.set_value(i, 'Notes', bracket)
    string = re.sub('\[.*?\]', '', string).strip() # Remove bracketed items
    df.set_value(i, 'URLs', string)
    print(string.split(' '))
    strings = string.split(' ')
    for url in strings: 
        if url.startswith('http') and url.endswith('txt'): 
            df.set_value(i, 'txtURL', url)
        if url.startswith('http') and ( url.endswith('html') or url.endswith('htm') ): 
            df.set_value(i, 'htmlURL', url)
```

```python
for i, row in df.iterrows(): 
    titleAuthor = row['Title and Author']
    elems = titleAuthor.split(',')
    title = ','.join(elems[:-1])
    author = elems[-1]
    author = author.strip()
    if author.startswith('by '): 
        author = author.replace('by ', '')
    if type(row.Author) != str: #NaN
        df.set_value(i, 'Author', author)
    else: 
        df.set_value(i, 'AuthorAlt', author)
    if type(row.Title) != str: #NaN
        df.set_value(i, 'Title', title)
    else: 
        df.set_value(i, 'ShortTitle', title)
    print(title, author)
```

```python
for i, row in df.iterrows(): 
    if type(row.txtURL) == str: 
        print('Getting text from: ', row.txtURL)
        if type(row.Text) != str: 
            r = requests.get(row.txtURL)
            if r.status_code == 200: 
                text = r.text
                df.set_value(i, 'Text', text)
                print('Processed successfully!')
            else: 
                print('Got bad response: ', r.status_code)
```
